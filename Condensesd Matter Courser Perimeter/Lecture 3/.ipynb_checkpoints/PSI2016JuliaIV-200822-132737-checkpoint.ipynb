# IJulia IV: free fermions

### Quadratic fermionic Hamiltonian

```julia
using PyPlot

N=11 # N odd work well, because it avoids accidental degeneracies in C*Tf
bc = 1 # boundary conditions
h=1. # criticality for h=1 

function buildC(N,h,bc) # choose bc=1 (default) for PBC and bc=-1 for APBC
    Z = zeros(2N,2N) # number operator
    J = zeros(2N,2N)
    for i=1:N
        Z[2i-1:2i,2i-1:2i] = [0 2h; -2h 0]
    end
    for i=1:N-1
        J[2i,2i+1] = 2
        J[2i+1,2i] = -2
    end
    J[1,2N] = -2*bc
    J[2N,1] = 2*bc
    return C = Z+J
end

C = buildC(N,h,1)
```

### Fermionic translation operator Tf

```julia
# Translation operator Tf

function buildTf(N,bc) # choose bc=1 (default) for PBC and bc=-1 for APBC
    Tf = zeros(2N,2N)
    for i=1:2N-2
        Tf[i,i+2] = 1
    end
    Tf[2N-1,1] = bc
    Tf[2N,2] = bc
    return(Tf)
end

Tf = buildTf(N,1)
```

```julia
# Test: do C and Tf commute?
C = buildC(N,h,1)
Tf = buildTf(N,1)
isequal(C*Tf,Tf*C)
```

### Simultaneous diagonalization of  Hamiltonian C and translation operator Tf
### Single particle energies $e_\alpha$ and momenta $k_{\alpha}$

```julia
C1 = buildC(N,1,1)  # periodic boundary conditions PBC
Tf1 = buildTf(N,1)
D,U = eig(C1*(eye(2N)+0.13*Tf1))
e1 = real(diag(im*U'*C1*U))
k1 = angle(diag(U'*Tf1*U))

plot(k1,e1, marker = "o", linestyle = "", color = "b")
```

```julia
perm = sortperm(e1) # let us sort out the single particle energies
e1 = e1[perm]
k1 = k1[perm]

e1 = e1[N+1:2N] # keep positive energies only
k1 = k1[N+1:2N]

plot(k1,e1, marker = "o", linestyle = "", color = "b")
 
[e1 k1]
```

```julia
C2 = buildC(N,1,-1) # anti-periodic boundary conditions APBC
Tf2 = buildTf(N,-1)
D,U = eig(C2*(eye(2N)+0.13*Tf2))
e2 = real(diag(im*U'*C2*U)) # single particle energies
k2 = angle(diag(U'*Tf2*U)) # single particle momentum

perm = sortperm(e2) # let us sort out the energies
e2= e2[perm]
k2 = k2[perm]

e2 = e2[N+1:2N] # keep positive energies only
k2 = k2[N+1:2N]

plot(k1,e1, marker = "o", linestyle = "", color = "b")
plot(k2,e2, marker = "o", linestyle = "", color = "r")

[e1 k1 e2 k2]
```
